```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
uploaded_filename = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[uploaded_filename]))
display(df)
```

```python
weight = df['weight'].values
height = df['height'].values/100

bmi_check = (weight)/(height**2)

df['Overweight'] = (bmi_check > 25).astype(int)

display(df)
```

```python
df['active'] = 1 - df['active']
display(df)
```

```python
chol_check = df['cholesterol'].values
df['cholesterol'] = (chol_check > 1).astype(int)
display(df)
```

```python
gluc_check = df['gluc'].values
df['gluc'] = (gluc_check > 1).astype(int)
display(df)
```

```python
df['age'] = df['age']/365
display(df)
```

```python
def draw_cat_plot(df):
  df = df.melt(id_vars=['cardio'], value_vars=['active', 'alco', 'cholesterol', 'gluc', 'Overweight'])
  fig = sns.catplot(data=df, kind='box', x='cardio', y='value', hue='variable')
  display(fig.fig)
```

```python
draw_cat_plot(df)
```

```python
def draw_heat_map(df):
  #Clean the data
  df = df[(df['height'] >= df['height'].quantile(0.025)) & (df['height'] <= df['height'].quantile(0.975))]
  df = df[(df['weight'] >= df['weight'].quantile(0.025)) & (df['weight'] <= df['weight'].quantile(0.975))]
  df = df[(df['ap_lo'] <= df['ap_hi'])]
  df = df[(df['ap_hi'] >= df['ap_lo'])]

  # Calculate the correlation matrix
  corr_matrix = df.corr()

  # Create a heatmap
  plt.figure(figsize=(12, 10))
  sns.heatmap(corr_matrix, annot=True, fmt=".1f", cmap="coolwarm", linewidths=.5)
  plt.title('Correlation Heatmap of Medical Examination Data')
  plt.show()
```

```python
draw_heat_map(df)
```
