```python
import numpy as np
import pandas as pd
```

```python
import io
uploaded_filename = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[uploaded_filename]))
display(df)
```

```python
df.describe()
```

```python
display(df.dtypes)
```

```python
# Percentage representation of races in dataframe
total_people = df.shape[0]
race_counts = df['race'].value_counts()
race_percentages = (race_counts / total_people) * 100
display(race_percentages)
```

```python
# Average age of men
df_filtered = df[df['sex'] == 'Male'].copy()
average_age_men = df_filtered['age'].mean()
print(f"The average age of men is {average_age_men:.2f} years.")
```

```python
# Percentage of people with a Bachelor's Degree
df_filtered = df[df['education'] == 'Bachelors'].copy()
total_people = df.shape[0]
bachelors_count = df_filtered.shape[0]
bachelors_percentage = (bachelors_count / total_people) * 100
print(f"{bachelors_percentage:.2f}% of people possess a Bachelor's Degree.")
```

```python
# Percentage of people with advanced education
df_filtered = df[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])].copy()
total_people = df.shape[0]
advanced_education_count = df_filtered.shape[0]
advanced_education_percentage = (advanced_education_count / total_people) * 100
print(f"{advanced_education_percentage:.2f}% of people have advanced education.")
```

```python
# Percentage of people with advanced education, that are earning over 50K
df_filtered_1 = df[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])].copy()
df_filtered_2 = df_filtered_1[df_filtered_1['salary'] == '>50K']
total_people = df_filtered_1.shape[0]
advanced_education_count = df_filtered_2.shape[0]
advanced_education_percentage = (advanced_education_count / total_people) * 100
print(f"{advanced_education_percentage:.2f}% of people with advanced education earn over 50K.")
```

```python
# Percentage of people without an advanced education, earning over 50K
df_filtered_1 = df[~df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])].copy()
df_filtered_2 = df_filtered_1[df_filtered_1['salary'] == '>50K']
total_people = df_filtered_1.shape[0]
advanced_education_count = df_filtered_2.shape[0]
advanced_education_percentage = (advanced_education_count / total_people) * 100
print(f"{advanced_education_percentage:.2f}% of people without advanced education earn over 50K.")
```

```python
# Minimum amount of hours worked
min_hours = df['hours-per-week'].min()
print(f"The minimum amount of time worked is {min_hours} hour(s).")
```

```python
# Percentage of people working minumum hours AND earning above 50K
df_filtered_1 = df[df['hours-per-week'] == min_hours].copy()
df_filtered_2 = df_filtered_1[df_filtered_1['salary'] == '>50K']
total_people = df_filtered_1.shape[0]
min_hours_count = df_filtered_2.shape[0]
min_hours_percentage = (min_hours_count / total_people) * 100
print(f"{min_hours_percentage}% of people work minimum hours and still earn above 50K")
```

```python
# Country with the highest number of people earning above 50K
df_filtered = df[df['salary'] == '>50K'].copy()
country_counts = df_filtered['native-country'].value_counts()
total_people = df_filtered.shape[0]
highest_earning_country = country_counts.idxmax()
highest_earning_country_percentage = (country_counts.max() / total_people) *100
print(f"{highest_earning_country} is the country with the highest number of people earning above 50K, with a percentage of {highest_earning_country_percentage:.2f}%")
```

```python
# The most popular job in India, for those earning above 50K
df_filtered = df[(df['native-country'] == 'India') & (df['salary'] == '>50K')].copy()
popular_job = df_filtered['occupation'].value_counts().idxmax()
print(f"The most popular job in India for those earning above 50K is {popular_job}.")
```
