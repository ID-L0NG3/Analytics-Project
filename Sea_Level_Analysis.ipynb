```python
# Filter the DataFrame for data from 2020 onwards
df_after_2019 = df[df['Year'] >= 2020].copy()

# Drop rows with NaN values in 'NOAA Adjusted Sea Level' for the new range
df_after_2019_cleaned = df_after_2019.dropna(subset=['NOAA Adjusted Sea Level'])

# Define the new range of x-values and y-values using the cleaned filtered data
new_x_values = df_after_2019_cleaned['Year'].values
new_y_values = df_after_2019_cleaned['NOAA Adjusted Sea Level'].values

# Perform linear regression on the filtered data
# Ensure there are enough data points for regression
if len(new_x_values) > 1:
    slope, intercept, r_value, p_value, std_err = linregress(new_x_values, new_y_values)

    # Define the limits for the regression line plot
    # Use the range of the new x-values
    regression_line_x = np.arange(new_x_values.min(), 2050)
    regression_line_y = slope * regression_line_x + intercept

    # plot the scatter graph
    plt.figure(figsize=(10, 6))
    plt.scatter(new_x_values, new_y_values, label='Original Data from the Year 2020')

    # plot the regression line
    plt.plot(regression_line_x, regression_line_y, color='red', label=f'Regression Line (Slope: {slope:.2f})')

    plt.xlabel('Year')
    plt.ylabel('Sea Level (inches)')
    plt.title('Rise in Sea Level')
    plt.legend()
    plt.grid(True)
    plt.show()
else:
    print("Not enough data points from 2020 onwards to perform regression and plotting.")
```

```python
sea_level_graphs = plt.subplots(nrows = 2, ncols = 2, figsize = (12, 10)) # Increased figsize for better readability
fig, ((ax1, ax2),(ax3, ax4)) = sea_level_graphs

dropped_df = df.dropna(subset=['CSIRO Adjusted Sea Level'])
xvalues = dropped_df['Year'].values
yvalues = dropped_df['CSIRO Adjusted Sea Level'].values

# Plotting a scatter graph of CSIRO Adjusted Sea Level against Year (In top-left
# subplot)
ax1.scatter(xvalues, yvalues)
ax1.set_ylabel('CSIRO Adjusted Sea Level (inches)')
ax1.set_xlabel('Year')
ax1.set_title('CSIRO Adjusted Sea Level vs Year')

# Plotting a regression line for the graph above on ax1
slope, intercept, r_value, p_value, std_err = linregress(xvalues, yvalues)
regression_line = slope * xvalues + intercept
ax1.plot(xvalues, regression_line, color='red', label=f'Regression Line (Slope: {slope:.2f})')
ax1.legend()

# Plotting an extended regression line on the second subplot (top-right)
extended_years = np.arange(xvalues.min(), 2051)
extended_regression_line = slope * extended_years + intercept
ax2.plot(extended_years, extended_regression_line, color='green', label=f'Extended Regression Line')
ax2.set_ylabel('CSIRO Adjusted Sea Level (inches)')
ax2.set_xlabel('Year')
ax2.set_title('Extended Regression Line')
ax2.legend()

# Plotting a graph of NOAA against Year, for all values including and after the
# year 2020. (bottom-left)
df_after_2019 = df[df['Year'] >= 2020].copy()
df_after_2019_cleaned = df_after_2019.dropna(subset=['NOAA Adjusted Sea Level'])
noaa_x_values = df_after_2019_cleaned['Year'].values
noaa_y_values = df_after_2019_cleaned['NOAA Adjusted Sea Level'].values

ax3.scatter(noaa_x_values, noaa_y_values)
ax3.set_ylabel('NOAA Adjusted Sea Level (inches)')
ax3.set_xlabel('Year')
ax3.set_title('NOAA Adjusted Sea Level from 2020')

# Plotting a regression line to predict the sea level in 2050, using the filtered
# dataframe on the bottom-right subplot (ax4)
if len(noaa_x_values) > 1:
    slope_noaa, intercept_noaa, r_value_noaa, p_value_noaa, std_err_noaa = linregress(noaa_x_values, noaa_y_values)

    # Define the limits for the regression line plot
    regression_line_x_noaa = np.arange(noaa_x_values.min(), 2051) # Extend to 2050
    regression_line_y_noaa = slope_noaa * regression_line_x_noaa + intercept_noaa

    ax4.scatter(noaa_x_values, noaa_y_values, label='Original Data from 2020')
    ax4.plot(regression_line_x_noaa, regression_line_y_noaa, color='red', label=f'Regression Line (Slope: {slope_noaa:.2f})')

    # Predict and plot the point for 2050
    predicted_sea_level_noaa_2050 = slope_noaa * 2050 + intercept_noaa
    ax4.plot(2050, predicted_sea_level_noaa_2050, color='black', marker='x', markersize=8, label=f'Predicted 2050: {predicted_sea_level_noaa_2050:.2f}')

    ax4.set_ylabel('Sea Level (inches)')
    ax4.set_xlabel('Year')
    ax4.set_title('NOAA Sea Level Rise from 2020 with Regression Line and 2050 Prediction')
    ax4.legend()
else:
    ax4.text(0.5, 0.5, "Not enough data points from 2020 onwards for NOAA regression", horizontalalignment='center', verticalalignment='center', transform=ax4.transAxes)


# Add tight_layout for better spacing
plt.tight_layout()
plt.show()
```
